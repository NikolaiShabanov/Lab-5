Варіант №8

\begin{equation*}
 \begin{cases}
   13.1x_1 + 9.2x_2 = 173\\
   8.3x_1 + 21x_2 = 271\\
   3.2x_1 - 8.3x_2 = 34\\
 \end{cases}
\end{equation*}

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

var1 = np.linspace(-10, 10, 100)
var2 = np.linspace(0, 25, 100)

y1 = (173 - 13.1*var1) / 9.2
y2 = (271 - 8.3*var1) / 21
y3 = (34 - 3.2*var1) / -8.3

fig = plt.figure(figsize=(15, 10))

gs = gridspec.GridSpec(3, 3, figure=fig)

plt.suptitle('The example')

# First chart
ax1 = plt.subplot(gs[0, 0])
ax1.plot(var1, y1, 'r--')
ax1.set_title('$13.1x_1 +9.2x_2 = 173$')
ax1.grid(True)

# Second chart
ax2 = plt.subplot(gs[0, 1])
ax2.plot(var1, y2, 'b:')
ax2.set_title('$8.3x_1 + 21x_2 = 271$')
ax2.grid(True)
ax2.set_yticklabels([])

# Third chart
ax3 = plt.subplot(gs[0, 2])
ax3.plot(var1, y3, 'm-.')
ax3.set_title('$3.2x_1 - 8.3x_2 = 34$')
ax3.grid(True)
ax3.set_yticklabels([])

y1 = (173 - 13.1 * var2) / 9.2
y2 = (271 - 8.3 * var2) / 21
y3 = (34 - 3.2 * var2) / -8.3

# Fourth chart
combined_ax = plt.subplot(gs[1:, :])
combined_ax.plot(var2, y1, 'r--', label='f[0]')
combined_ax.plot(var2, y2, 'b:', label='f[1]')
combined_ax.plot(var2, y3, 'm-.', label='f[2]')
combined_ax.set_title('All lines in one figure')
combined_ax.legend(loc="upper right")
combined_ax.grid(which='major', color='g', linestyle='-', linewidth=0.3)
combined_ax.set_xlabel("$x_1$")
combined_ax.set_ylabel("$x_2$")
combined_ax.set_xlim([0, 25])
combined_ax.set_ylim([-5, 15])

# Dots
A1 = np.array([[13.1, 9.2], [8.3, 21]])
A2 = np.array([[13.1, 9.2], [3.2, -8.3]])
A3 = np.array([[8.3, 21], [3.2, -8.3]])

B1 = np.array([173, 271])
B2 = np.array([173, 34])
B3 = np.array([271, 34])

x1 = np.linalg.solve(A1, B1)
x2 = np.linalg.solve(A2, B2)
x3 = np.linalg.solve(A3, B3)

X = np.array([[x1[0], x1[1]], [x2[0], x2[1]], [x3[0], x3[1]]])
t1 = plt.Polygon(X, color='lightgray')
combined_ax.add_patch(t1)

combined_ax.plot(x1[0], x1[1], 'o', color="cyan")
combined_ax.plot(x2[0], x2[1], 'o', color="orange")
combined_ax.plot(x3[0], x3[1], 'o', color="purple")

# Lettering and arrows
combined_ax.text(8, 3, '$13.1x_1 +9.2x_2 = 173$', rotation=-39)
combined_ax.text(10, 8, '$8.3x_1 + 21x_2 = 271$', rotation=-13)
combined_ax.text(4, -2.3, '$3.2x_1 - 8.3x_2 = 34$', rotation=11)

combined_ax.arrow(x=12.7, y=0.8, dx=-1, dy=-2, width=.09)
combined_ax.text(8, -3, 'The intersection point of f[0] and f[2]', rotation=0)

combined_ax.arrow(x=21.8, y=4.1, dx=0, dy=-2, width=.08)
combined_ax.text(19, 1, 'The intersection point of f[1] and f[2]', rotation=0)

combined_ax.arrow(x=5.8, y=10.8, dx=2, dy=2, width=.08)
combined_ax.text(6, 14, 'The intersection point of f[0] and f[1]', rotation=0)

# Save the figure
fig.savefig("filename1.jpg", dpi=200)
fig.savefig("filename2.png", dpi=200)
fig.savefig("filename3.svg", dpi=200)
```
